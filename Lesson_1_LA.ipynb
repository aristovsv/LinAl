__1.__ –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –Ω–∞ –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

–í –¥–∞–Ω–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –ª–∏–Ω–µ–π–Ω–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å, —Ç–∞–∫ –∫–∞–∫ –º—ã –º–æ–∂–µ–º –≤—ã—Ä–∞–∑–∏—Ç—å $$f_{4}(x) = f_{3}(x) - f_{1}(x) - f_{2}(x)$$

__2.__ –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –Ω–∞ –ª–∏–Ω–µ–π–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

–í –¥–∞–Ω–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ —Ç–∞–∫–∂–µ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –ª–∏–Ω–µ–π–Ω–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å, —Ç–∞–∫ –∫–∞–∫ –º—ã –º–æ–∂–µ–º –≤—ã—Ä–∞–∑–∏—Ç—å $$f_{4}(x) = f_{3}(x) + 2f_{2}(x) + f_{1}(x) * 0,5$$

$$f_{4}(x) = x^2 + 2x + 2 * 0,5$$

__3.__ –ù–∞–π–¥–∏—Ç–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤–µ–∫—Ç–æ—Ä–∞ $x = (2, 3, 5)\in \mathbb{R}^{3}$ –≤ –±–∞–∑–∏—Å–µ $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$$\alpha_{1}(0, 0, 10) + \alpha_{2}(2, 0, 0) + \alpha_{3}(0, 1, 0) = (0, 0, 0)$$
–†–µ—à–∞—è —ç—Ç—É —Å–∏—Å—Ç–µ–º—É, –ø–æ–ª—É—á–∏–º Œ±1 = Œ±2 = Œ±3 = 0, —Ç. –µ. –≤–µ–∫—Ç–æ—Ä—ã b1, b2, b3 –ª–∏–Ω–µ–π–Ω–æ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã, –ø–æ—ç—Ç–æ–º—É –æ–Ω–∏ –æ–±—Ä–∞–∑—É—é—Ç –±–∞–∑–∏—Å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ R^3.

–±) –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã, —Å—á–∏—Ç–∞—é—â–∏–µ—Å—è –ª–∏–Ω–µ–π–Ω—ã–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –¥–∞–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ $\{u_{1}, u_{2} \ldots, u_{n}\}$.

__1.__ –ù–∞–π–¥–∏—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ $x, y \in \mathbb{R}$:<br>

```python
x = np.array([7, -4, 0, 1])
y = np.array([-3, 1, 11, 2])
print(f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ x –∏ y:\n{np.dot(x, y)}')
```

__2.__ –ù–∞–π–¥–∏—Ç–µ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ $(4, 2, 4)$ –∏ $(12, 3, 4)$ –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

```python
a = np.array([4, 2, 4])
b = np.array([12, 3, 4])

print(f'l1 –ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ a:\n{norm(a, ord=1)}')
print(f'l2 –ï–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ –∞:\n{norm(a, ord=2)}')

print(f'l1 –ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ b:\n{norm(b, ord=1)}')
print(f'l2 –ï–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ b:\n{norm(b, ord=2)}')
```

```python
cos_phi = np.dot(a, b) / norm(a) / norm(b)
print(f'–ö–æ—Å–∏–Ω—É—Å —É–≥–ª–∞: {cos_phi}')
```

__3.__ –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –±—É–¥–µ—Ç –ª–∏ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –µ–≤–∫–ª–∏–¥–æ–≤—ã–º, –µ—Å–ª–∏ –∑–∞ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –ø—Ä–∏–Ω—è—Ç—å:<br>
–∞) –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–ª–∏–Ω –≤–µ–∫—Ç–æ—Ä–æ–≤;<br>
–±) —É—Ç—Ä–æ–µ–Ω–Ω–æ–µ –æ–±—ã—á–Ω–æ–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤?<br>

a) –õ–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –Ω–∞–¥ –ø–æ–ª–µ–º –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è _–µ–≤–∫–ª–∏–¥–æ–≤—ã–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º_, –µ—Å–ª–∏ –Ω–∞ –Ω—ë–º –≤–≤–µ–¥–µ–Ω–æ –ø—Ä–∞–≤–∏–ª–æ, —Å–æ–≥–ª–∞—Å–Ω–æ –∫–æ—Ç–æ—Ä–æ–º—É –∫–∞–∂–¥–æ–π —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω–æ–π –ø–∞—Ä–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ $x$ –∏ $y$ –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–æ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ, –Ω–∞–∑—ã–≤–∞–µ–º–æ–µ _—Å–∫–∞–ª—è—Ä–Ω—ã–º –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ–º_ –∏ –æ–±–æ–∑–Ω–∞—á–∞–µ–º–æ–µ —Å–∏–º–≤–æ–ª–æ–º $(x, y)$.

–£–∫–∞–∑–∞–Ω–Ω–æ–µ –ø—Ä–∞–≤–∏–ª–æ –¥–æ–ª–∂–Ω–æ –ø–æ–¥—á–∏–Ω—è—Ç—å—Å—è —á–µ—Ç—ã—Ä—ë–º –∞–∫—Å–∏–æ–º–∞–º:<br>

b) –Ø –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—é, —á—Ç–æ –¥–∞–Ω–Ω–æ–µ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ —Ç–∞–∫–∂–µ —è–≤–ª—è–µ—Ç—Å—è –µ–≤–∫–ª–∏–¥–æ–≤—ã–º, –¥–∞–Ω–Ω—ã–µ 4 –∞–∫—Å–∏–æ–º—ã –±—É–¥—É—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è.

__4.__ –í—ã—è—Å–Ω–∏—Ç–µ, –∫–∞–∫–∏–µ –∏–∑ –Ω–∏–∂–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ $\mathbb{R}^{3}$:<br>

a) $(1,0,0),(0,0,1);$ <br>
–Ω–µ —è–≤–ª—è–µ—Ç—Å—è –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–º, —Ç–∞–∫ –∫–∞–∫ –ø—Ä–æ—Å—Ç—Ä–∞–Ω–Ω–æ—Å—Ç–≤–æ $\mathbb{R}^{3}$

–±) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
